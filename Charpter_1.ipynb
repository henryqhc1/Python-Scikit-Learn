```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

plt.rcParams['figure.figsize'] = [8,6]
sns.set_style("darkgrid")

titanic_data=sns.load_dataset('titanic')

titanic_data = titanic_data[['age','fare','pclass']]
```

```
titanic_data.describe()
```

```
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

scaler.fit(titanic_data)

titanic_data_scaled=scaler.transform(titanic_data) # result is an array
```

```
titanic_data_scaled=pd.DataFrame(titanic_data_scaled, columns=titanic_data.columns)
titanic_data_scaled.head()
```

```
sns.kdeplot(titanic_data['age'])
```

```
sns.kdeplot(titanic_data_scaled['age'])
```

```
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()

scaler.fit(titanic_data)

titanic_data_scaled=scaler.transform(titanic_data) # result is an array

titanic_data_scaled=pd.DataFrame(titanic_data_scaled, columns=titanic_data.columns)
titanic_data_scaled.head()
```

```
sns.kdeplot(titanic_data['age'])
```

```
sns.kdeplot(titanic_data_scaled['age'])
```

```
from sklearn.preprocessing import Normalizer

scaler=Normalizer()

scaler.fit(titanic_data)

titanic_data_scaled=scaler.transform(titanic_data) # result is an array

titanic_data_scaled=pd.DataFrame(titanic_data_scaled, columns=titanic_data.columns)
titanic_data_scaled.head()
```

```
titanic_data.isnull().mean()
```

```
print('age median',titanic_data.age.median())

print('age mean',titanic_data.age.mean())
```

```
titanic_data['Median_age']=titanic_data['age'].fillna(np.round(titanic_data.age.median()))

titanic_data['Mean_age']=titanic_data['age'].fillna(np.round(titanic_data.age.mean()))

titanic_data.head(20)
```
